# Reviewer replication path

This notebook is a paper-style companion to `docs/REVIEWER_GUIDE.md`. It does not ask you to trust prose. It loads the frozen reviewer evidence manifest, checks the closure ledger, inspects the frozen MPC-derived labelled corpus manifest, and shows the one-command audit path.

````python
from pathlib import Path
import json

ROOT = Path('..').resolve() if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
manifest_path = ROOT / 'data/benchmarks/reviewer_evidence/reviewer_evidence_manifest.json'
manifest = json.loads(manifest_path.read_text(encoding='utf-8'))
manifest['status'], manifest['file_count'], manifest['certificate_hash']
````

## Claim-to-artifact table

Every row below is hashed in the reviewer evidence manifest. If any tracked artifact changes, the manifest certificate changes too.

````python
for rec in manifest['files']:
    print(f"{rec['role']:<34} {rec['path']:<82} {rec['sha256'][:16]}")
````

## Closure ledger

The closure ledger is the compact pass/fail readiness object used by the public launch audit.

````python
closure = json.loads((ROOT / 'data/benchmarks/closure/closure_report.json').read_text(encoding='utf-8'))
{
    'status': closure['status'],
    'readiness_score': closure['readiness_score'],
    'critical_failures': closure['critical_failures'],
    'certificate_hash': closure['certificate_hash'],
}
````

## Frozen labelled corpus

This is the existing MPC-derived 500-case labelled corpus. It is not a blind NOIRLab/ZTF/LSST benchmark, but it is a frozen, hash-addressed corpus with benchmark outputs and calibration artifacts.

````python
corpus = json.loads((ROOT / 'data/benchmarks/real_corpus_mpc_500/corpus_manifest.json').read_text(encoding='utf-8'))
{
    'schema': corpus['schema'],
    'n_cases': corpus['n_cases'],
    'case_hash': corpus['case_hash'],
    'benchmark': corpus['benchmark'],
}
````
